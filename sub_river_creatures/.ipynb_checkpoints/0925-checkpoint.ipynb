```python
import pandas as pd
```

```python
df_all=pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\工作\eDNA\sub_river_creatures\所有物種.csv')
df_F=pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\工作\eDNA\sub_river_creatures\夫布爾溪.csv')
df_M=pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\工作\eDNA\sub_river_creatures\米磨登溪.csv')
df_B=pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\工作\eDNA\sub_river_creatures\寶谷富溪.csv')

df_all=df_all.values.tolist()
df_F=df_F.values.tolist()
df_M=df_M.values.tolist()
df_B=df_B.values.tolist()
```

```python
print(len(df_all))
print(len(df_F))
print(len(df_M))
print(len(df_B))
```

```python
list_F=[]
for i in range(len(df_all)):
    if(df_all[i] in df_F):
        #print(df_all[i])
        list_F.append(1)
    else:
        list_F.append(0)

list_M=[]
for i in range(len(df_all)):
    if(df_all[i] in df_M):
        #print(df_all[i])
        list_M.append(1)
    else:
        list_M.append(0)
        
list_B=[]
for i in range(len(df_all)):
    if(df_all[i] in df_B):
        #print(df_all[i])
        list_B.append(1)
    else:
        list_B.append(0)
```

```python
print(len(df_all))

print(list_F)
print(len(list_F))

print(list_M)
print(len(list_M))

print(list_B)
print(len(list_B))
```

```python
# dictionary of lists
dict = {'all': df_all, 'F': list_F, 'M': list_M, 'B': list_B}
     
df = pd.DataFrame(dict)
     
# saving the dataframe
df.to_csv('GFG.csv')
```
